# Data Preprocessing

```python
# Drop Highly Corelated column
data = data.drop(columns=['maxtemp', 'temparature', 'mintemp'])
```

```python
data.head()
```

```python
# Finding Majority and Minority class
data["rainfall"].value_counts()
```

```python
# Separating Majority and Minority classes
df_majority = data[data["rainfall"] == 1]
df_minority = data[data["rainfall"] == 0]
```

```python
print(df_majority.shape)
print(df_minority.shape)
```

```python
# Downsampling majority class to match minority class
df_majority_downsampled = resample(df_majority, replace=False, n_samples=len(df_minority), random_state=42)
```

```python
df_majority_downsampled.shape
```

```python
df_downsampled = pd.concat([df_majority_downsampled, df_minority])
```

```python
df_downsampled.shape
```

```python
df_downsampled.head()
```

```python
# Shuffling the final dataframe
df_downsampled = df_downsampled.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
df_downsampled.head()
```

# Spliting the Data into Training and Testing

```python
# split the features and target in x and y
X=df_downsampled.drop(columns=["rainfall"])
Y=df_downsampled["rainfall"]
```

```python
print(X)
print(Y)
```

```python
(X_train, X_test, Y_train, Y_test) = train_test_split(X, Y, test_size=0.2, random_state=42)
```

# Model Training

```python
# Defing the hyperparameters of random forest classifier using GridSearchCV
rf_model = RandomForestClassifier(random_state=42)

param_grid_rf = {
    "n_estimators": [50, 100, 200],
    "max_features": ["sqrt", "log2"],
    "max_depth": [None, 10, 20, 30],
    "min_samples_split":[2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}
```

```python
# Hyperparameter Tuning using GridSearchCV

grid_search_rf = GridSearchCV(estimator=rf_model, param_grid=param_grid_rf, cv=5, n_jobs=-1, verbose=2)
grid_search_rf.fit(X_train, Y_train)
```

```python
best_rf_model = grid_search_rf.best_estimator_
print("Best Parameters for Random Forest:", grid_search_rf.best_params_)
```

# Model Evaluation

```python
cv_scores = cross_val_score(best_rf_model, X_train, Y_train, cv=5)
print("Cross Validation Scores:",cv_scores)
print("Mean Cross Validation Scores:",np.mean(cv_scores))
```

```python
# Test set performance
y_pred = best_rf_model.predict(X_test)

print("Test Set Accuracy:", accuracy_score(Y_test, y_pred))
print("Test Set of Confusion Matrix:\n", confusion_matrix(Y_test, y_pred))
print("Classification Report:\n", classification_report(Y_test, y_pred))
```
